```python
import yt
```

```python
ds_cyl = yt.load("data_cyl_test")
print(ds_cyl.geometry)
```

```python
s_cyl = yt.SlicePlot(ds_cyl, "theta", ("boxlib", "phi"))
s_cyl.show()
s_cyl.save("cylindrical.png")
```

```python
data_cyl = ds_cyl.all_data()
fmin, fmax = data_cyl.quantities.extrema(("boxlib", "phi"))
print("min: ", fmin)
print("max: ", fmax)
```
